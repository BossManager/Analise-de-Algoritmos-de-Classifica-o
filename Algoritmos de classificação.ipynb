```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.preprocessing import Normalizer,MinMaxScaler
import numpy as np
def scores(y_test,prediction):
    print("Accuracy: " + str(accuracy_score(y_test, prediction) * 100) + "%")
    print("Precision: " + str(precision_score(y_test, prediction,average='micro') * 100) + "%")
    print("F1 score: " + str(f1_score(y_test, prediction,average='micro') * 100) + "%")
    print("Recall: " + str(recall_score(y_test, prediction,average='micro') * 100) + "%")
```

# Leitura do dataset: winequality

```python
winered = pd.read_csv("winequality-red.csv",sep=";")
winewhite = pd.read_csv("winequality-white.csv",sep=";")
```

<h2>Adicionando o tipo de vinho a tabela</h2>

```python
winered['type'] = 0 #red=0
winewhite['type'] = 1#white=1
```

```python
winered.head()
```

```python
#concatenando as duas tabelas e embaralhando as linhas
#wines = pd.concat([winered,winewhite]).sample(frac=1).reset_index(drop=True)
X_train, X_test, y_train, y_test = train_test_split(winered.iloc[:,0:11], winered.iloc[:,-1], test_size=0.2)
```

```python
#Normalizando os dados
# Create normalizer
normalizer = Normalizer(norm='l2')#max l1 l2
scaller = MinMaxScaler(feature_range=(0, 12))
# Transform feature matrix
X_train = normalizer.transform(X_train)
X_test = normalizer.transform(X_test)
#X_train = scaller.fit_transform(X_train)
#X_test = scaller.fit_transform(X_test)
```

<h2>Decision Tree</h2>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create decision tree classifer object using gini
clf = DecisionTreeClassifier(criterion='gini', random_state=0)
# Train model
model = clf.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
classes =[str(x) for x in list(model.classes_)]
```

```python
from sklearn import datasets
from IPython.display import Image  
from sklearn import tree
import pydotplus
```
